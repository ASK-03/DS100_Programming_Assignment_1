```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.util import random_noise
from PIL import Image
```

```python
img = cv2.imread('scenery.jpg', 1)
gauss = np.asarray(random_noise(img, mode='gaussian', seed=None, clip=True))

plt.subplot(1,1,1), plt.imshow(gauss), plt.title('gaussian')
#plt.savefig('scenery_gaussian.png')
```
