```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from PIL import Image, ImageOps
```

```python
img = Image.open("scenery.jpg")
img_gray = ImageOps.grayscale(img)
image_Array_2D = np.asarray(img_gray)
```

```python
u, s, v = np.linalg.svd(image_Array_2D, full_matrices=False) 
#print(u.shape)
#print(s.shape)
#print(v.shape)
```

```python
components = [3614, 1, 5, 10, 20,50, 80, 100, 200]
plt.figure(figsize=(30, 15))

for i in range(len(components)):
    low_rank = u[:, :components[i]] @ np.diag(s[:components[i]]) @ v[:components[i], :]
    plt.subplot(3, 3, i+1),
    plt.imshow(low_rank, cmap='gray'),
    plt.axis('off'),
    plt.title(f'n_components = {components[i]}', )
    Image.fromarray(low_rank.astype(np.uint8)).save(f"img_compressed_using_SVD_{i}.jpg")
    #plt.savefig(f'img_compressed_scenery_{i}.pdf')
    #img_compressed = Image.fromarray(low_rank, 'I')
    #img_compressed.save(f'img_compressed_scenery_{i}.png')
    #img_compressed.show()
```
