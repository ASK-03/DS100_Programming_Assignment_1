```python
import matplotlib.image as mplib
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.decomposition import PCA
```

```python
img = mplib.imread('spiderman.jpg')
image_shape = img.shape
image_reshape = np.reshape(img, (image_shape[0], image_shape[1]*image_shape[2]))
```

```python
pca = PCA(500).fit(image_reshape)
img_trans = pca.transform(image_reshape)
```

```python
img_INV = pca.inverse_transform(img_trans)
img = np.reshape(img_INV, image_shape)
```

```python
plt.axis('off')
plt.imshow(img.astype('uint8'))
plt.show()
```
